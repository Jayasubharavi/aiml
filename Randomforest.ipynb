```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
```

```python
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([0, 0, 0, 1, 1, 1])

```

```python

model = RandomForestClassifier(n_estimators=10, random_state=42)
```

```python
model.fit(X, y)
```

```python
x_test = np.linspace(0, 7, 100).reshape(-1, 1)
y_prob = model.predict_proba(x_test)[:, 1]
```

```python
plt.scatter(X, y)  # actual data points
plt.plot(x_test, y_prob)  # probability curve
plt.xlabel("Hours Studied")
plt.ylabel("Probability of Passing")
plt.title("Random Forest Graph")
plt.show()
```

```python
print("Prediction for 3 hours:", model.predict([[3]]))
print("Prediction for 5 hours:", model.predict([[5]]))
```
